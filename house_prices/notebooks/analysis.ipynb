```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("../data/train.csv")

# See the first lines
df.head()
```

```python
# General Info
df.info()
```

```python
# Basics Statistics
df.describe()
```

```python
# See Null values
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
# Scatter plot between useful area and price
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["GrLivArea"], y=df["SalePrice"])
plt.title("Relationship between useful area and price")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", annot=False)
plt.title("Correlation Map")
plt.show()
```

```python
# Fill null values â€‹â€‹in a column with the mean
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].mean())

# Or, if you want to remove rows with null values:
# df = df.dropna(subset=['LotFrontage'])
```

```python
# See Null values
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
# Use charts like boxplots to see outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["SalePrice"])
plt.title("Boxplot of Sales Price")
plt.show()
```
